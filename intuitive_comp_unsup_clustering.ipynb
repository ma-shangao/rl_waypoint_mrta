# Intuitively comparison between the unsupervised minCUT clustering and k-means clustering

The comparison is run with the simplest 2D points data.

## Import preliminaries

```python
import numpy as np
from spek_gnn_clustering import GnnClustering
from utilities import plot_the_clustering_2d
from sklearn.cluster import k_means
```

## Generate data

In this instance, we set the cluster number as $k = 3$.

```python
X = np.random.rand(50, 2)
k = 3
```

## The minCUT clustering


```python
minCut = GnnClustering(k)

minCut.fit(X)
minCut.eval()
s_mc = minCut.s_out
plot_the_clustering_2d(k, s_mc, X)
```

## k-means clustering

```python
km_results = k_means(X, k)
s_km = km_results[1]
plot_the_clustering_2d(k, s_km, X)
```
